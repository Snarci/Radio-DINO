```python
from umap import UMAP
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#after running the feature extraction script, load the data
df = pd.read_csv('embeddings.csv')

#drop the filename column 
df = df.drop(columns=['filename'])

#use the label column as the y value
y = df['label']
#drop the label column
X = df.drop(columns=['label'])
```

```python
umap = UMAP(n_neighbors=15, min_dist=0.1, n_components=2, metric='cosine', random_state=42)
#fit the model
X_umap = umap.fit_transform(X)
#convert to a dataframe
X_umap = pd.DataFrame(X_umap, columns=['x', 'y'])
#add the label column
X_umap['label'] = y.values
#plot the data
plt.figure(figsize=(10, 10))
plt.scatter(X_umap['x'], X_umap['y'], c=X_umap['label'], cmap='Spectral', s=5)
plt.colorbar()
plt.title('UMAP projection of the dataset')
```
